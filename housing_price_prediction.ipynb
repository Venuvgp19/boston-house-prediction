```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housing.data',delim_whitespace=True, header = None)
```

```python
col_name = ['CRIM', 'ZN' , 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
df.columns = col_name
```

```python
df.describe()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

# Exploratory Data Analysis

```python
df.corr()
```

```python
col_study = ['RM','LSTAT','PTRATIO','MEDV']
```

```python
sns.pairplot(df[col_study],height = 1.5)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df[col_study].corr(),annot = True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression #importing Linear regression model 
X = df['RM'].values.reshape(-1,1) #training data
y = df['MEDV'].values # dependent variable
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.figure(figsize=(16,10))
sns.regplot(X,y) #plot
plt.xlabel('Number of rooms per dwelling')
plt.ylabel('House price in $1000\'s')
plt.show()
```

```python
model.predict(np.array([5]).reshape(-1,1)) #model prediction for 5 rooms roughly the pricing is 10.8*1000$
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
model3 = RANSACRegressor()
```

```python
model3.fit(X,y)
```

```python
inlier_mask = model3.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = model3.predict(line_X.reshape(-1, 1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,8));
plt.scatter(X[inlier_mask], y[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper left')
plt.show()
```
